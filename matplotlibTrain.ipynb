```python
import matplotlib.pyplot as plt  #matplotlib
import numpy as np  #numpy

#axisartist
import mpl_toolkits.axisartist as axisartist
#
fig = plt.figure(figsize=(6,4))
#axisartist.Subplotax
ax = axisartist.Subplot(fig, 111)  
#
fig.add_axes(ax)

x_bar=[(-1)*np.pi,(-1)*np.pi/2,np.pi/2,np.pi]
y_bar=[1,0,-1]
x_value=np.linspace(-1*np.pi,np.pi,1000)
y_value1=np.sin(x_value)
y_value2=np.cos(x_value)


ax.axis[:].set_visible(False)#set_visible
ax.axis["x"] = ax.new_floating_axis(0,0)#ax.new_floating_axis
ax.axis["x"].set_axisline_style("->", size = 1.0)#x

#y
ax.axis["y"] = ax.new_floating_axis(1,0)
ax.axis["y"].set_axisline_style("->", size = 1.0)
#xy
ax.axis["x"].set_axis_direction("bottom")
ax.axis["y"].set_axis_direction("left")

#x-y
#xy
xl, xr = -3.3,3.3
yu, xd = -1.1, 1.1
plt.xlim(xl, xr)
plt.ylim(yu, xd)
plt.xticks(x_bar)
plt.yticks(y_bar)

#name = r'$y=x(x-1)(x-2)$'
#plt.text(3.5, 6, name)
#plt.text(3, -0.1, 'x')
#plt.text(-0.1, 1.1, 'y')

plt.annotate('cos(*)=-',xy=(2*np.pi/3,-0.5),xytext=(0.3*np.pi,-0.8),arrowprops=dict(facecolor='black', width=1, headwidth=4))
plt.annotate('sin(*)=*1.74',xy=(2*np.pi/3,0.5*1.74),xytext=(0.82*np.pi,0.93),arrowprops=dict(facecolor='black', width=1, headwidth=4))
plt.hlines(0,-1*np.pi-0.2,np.pi,lw=1)
plt.vlines(0,-1.2,1.2,lw=1)
plt.vlines(2*np.pi/3,np.cos(2*np.pi/3),0,ls='--',color='b')
plt.vlines(2*np.pi/3,0,np.sin(2*np.pi/3),ls='--',color='r')

#
plt.plot(x_value,y_value2,color='b',lw='2')
plt.plot(x_value,y_value1,color='r',lw='2')
plt.legend(['Cos function','Sin Function'],loc=2)
```
